## Using Gaussian processes to evenly sample the data.

```python
from george import kernels
from george import GP
from scipy.optimize import minimize
from scipy.optimize import Bounds
```

Here I use the squared exponential kernel:
$$
k_{i, j}=A \exp \left[-\frac{\left(t_{i}-t_{j}\right)^{2}}{\lambda^{2}}\right]
$$

A is the amplitude and $\lambda$ is the decorrelation timescale. Metric = $\lambda^2$ 

```python
#Initial guesses for the hyperparameters
constant_guess = 2
metric_guess = 6

#Create the kernel
kernel = constant_guess * kernels.ExpSquaredKernel(metric =metric_guess)

#Initialize the GP
gp = GP(kernel, fit_kernel=True)

#define the objective function (negative log-likelihood)
def nll(p):
    gp.set_parameter_vector(p)
    ll=gp.log_likelihood(flux)
    return -ll if np.isfinite(ll) else 1e25

def grad_nll(p):
    gp.set_parameter_vector(p)
    ll = gp.grad_log_likelihood(flux)
    return -ll 

lower_bounds = [-20.,-2.]
upper_bounds = [20.,10.]
bounds = Bounds(lower_bounds,upper_bounds)


def GPpredict(mjd,flux):

    #Initial guesses for the hyperparameters
    constant_guess = 2
    metric_guess = 6

    #Create the kernel
    kernel = constant_guess * kernels.ExpSquaredKernel(metric =metric_guess)

    #Initialize the GP
    gp = GP(kernel, fit_kernel=True)
    
    #Compute the GP
    gp.compute(mjd)
    
    #find the best parameters
    result = minimize(nll, gp.get_parameter_vector(), method='L-BFGS-B',
              bounds=bounds)
    
    #pass the parameters to GP
    gp.set_parameter_vector(result.x)
    
    #recompute the GP
    gp.recompute()
    
    xpred = np.linspace(0,900,200)
    fpred,cov = gp.predict(flux,xpred)
    return xpred,fpred
```

```python
ndata = 200
nobjects = len(objects)

#create some arrays to store all of the samples.
mjd_all = np.zeros((6*nobjects,ndata))
flux_all = np.zeros((6*nobjects,ndata))
obj_all = np.zeros(nobjects)
mjd_all.shape
```

#### Run GP for the first 300 objects. 

```python
#It takes a very long time.
#run GP for all objects
for i, obj in enumerate(objects):
    for j in passbands:
        
        mjd = lcvs[lcvs['object_id'] == objects[i]][lcvs.passband==j].mjd.to_numpy()
        flux = lcvs[lcvs['object_id'] == objects[i]][lcvs.passband==j].flux.to_numpy()
        
        #set a common time frame and normilize the flux
        mjd = mjd -mjd[0]
        flux = (flux -flux.mean())/flux.std()
        
        gp.compute(mjd)
        new_mjd,new_flux = GPpredict(mjd,flux)

        mjd_all[i*6+j] = new_mjd
        flux_all[i*6+j] = new_flux
        
```

#### Check the results.

```python
#try some different n to see the results.
nobjs=[1,20,80,99]
for n in nobjs:
    plt.figure()
    plt.plot(mjd_all[n],flux_all[n],label='GP')
    i = n//6
    j=n%6

    mjd = lcvs[lcvs['object_id'] == objects[i]][lcvs.passband==j].mjd.to_numpy()
    flux =lcvs[lcvs['object_id'] == objects[i]][lcvs.passband==j].flux.to_numpy()
    mjd = mjd -mjd[0]
    flux = (flux -flux.mean())/flux.std()
    plt.plot(mjd,flux,'.',label='original data')

    plt.xlabel('time')
    plt.ylabel('flux')
    plt.legend();
```

## Autoencoder

```python
mjd_all.shape
```

- The input dimension is 200. Each bandpass is considered as a seperate time series.

- Optimization: Adamax. Adam is good in most cases. Adamax is a variant of Adam but more stable, especially to get rid of division by zero.
- loss: mse. mean squared error. Use mse for regression problems.
- Activation: ReLu. Rectified linear unit. It's cheap to compute and coverges fast. It's the most commonly used activation function in neural networks.

```python
model = Sequential()
model.add(Dense(128,activation='relu',input_dim=mjd_all.shape[1]))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation ='relu'))
model.add(Dense(16, activation ='relu'))
model.add(Dense(32, activation ='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(200,activation='relu'))

model.compile(optimizer='adamax',loss='mse')
print(model.summary())          
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(flux_all,flux_all,
                                                test_size=.25,random_state=1)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
#It takes a very long time.
history = model.fit(x_train, y_train, epochs=10000, batch_size=10, verbose=1)
```

```python
plt.plot(np.array(history.history['loss']))
#plt.yscale('log')
#plt.xscale('log')
plt.xlabel('iteration')
plt.ylabel('Loss');
```

**Figure 2:** The loss functions againest epochs. It tends to become flatten with 10000 epochs, so it should be good to use.

#### See the results from autoencoder.

```python
#randomly choose some objects:
for ii in [1,5,8,20,50]:

    ypred = model.predict(x_test)
    xpred = np.linspace(0,900,200)
    plt.figure()
    plt.plot(xpred,ypred[ii],label='decoded lightcurve')
    plt.plot(xpred,x_test[ii],alpha=0.5,label='original lightcurve')
    plt.xlabel('time')
    plt.ylabel('flux')
    plt.legend();
```

**Figure 3:** It shows some examples of the lightcurves generated by the autoencoder. There are some significant features remaining in the lightcurves but it loses some small features.

## Extract bottleneck features.

```python
def get_bottleneck_features(model,inputs):
    features = model.predict(inputs, verbose=0)
    return features
```

```python
import itertools
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    #print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, "%.2f"%cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)


    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    
# Compute confusion matrix
cnf_matrix = sklearn.metrics.confusion_matrix(y_testddf, clfddf.predict(X_testddf))

np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure(figsize=(10,10))

plot_confusion_matrix(cnf_matrix, classes=["%d"%i for i in np.unique(y)],
                      title='Confusion matrix, without normalization')

# Plot normalized confusion matrix
plt.figure(figsize=(10,10))
np.set_printoptions(precision=2)
plot_confusion_matrix(cnf_matrix, classes=["%d"%i for i in np.unique(y)], normalize=True,
                      title='Normalized confusion matrix');

plt.show();
```

**Figure:** It shows the Normalized and non-normalized comfusion matrices using random forest trees for test data of DDF data. The values mean the percentage of the the data that were correctly labelled.
